```python
#import the necessary packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read data file as pandas dataframe parsing the date and time columns 

SnowDepth = pd.read_csv('SnowEx2020_SnowDepths_COGM_alldepths_v01.csv', parse_dates= {'Datetime': [2,3]})
SnowDepth.rename(columns= {'Measurement Tool (MP = Magnaprobe; M2 = Mesa 2; PR = Pit Ruler)': 'Measurement Tool'}, inplace = True)
SnowDepth
```

```python
# Select records within the the period of GM IOP campaign
GMP_SnowDepth = SnowDepth[(SnowDepth['Datetime']>='1/28/2020') & (SnowDepth['Datetime'] <='2/12/2020')]
GMP_SnowDepth
```

```python
#Get information of the content of the dataframe
GMP_SnowDepth.info()
```

Que: How come the snowdepth is an int. The measurement precision of the instrument is only in cm? 

```python
# Summary statistics
GMP_SnowDepth[['Depth (cm)', 'elevation (m)']].describe()
```

How is the elevation in thousands of meter? There is NAN (-9999) in the elevation values.

```python
# group record by 'Measurement Tools' and show the {summary statistics} on the 'Snow Depth'. result is multi index dataframe

GMP_SnowDepth_MT = GMP_SnowDepth.groupby(['Measurement Tool'])[['Depth (cm)']].describe().round(1)
GMP_SnowDepth_MT.reset_index(inplace = True) #reset index to default, so that the groupby column (MT) is a column
GMP_SnowDepth_MT
```

```python
# Select records associated with a certain Measurement tool
GMP_SnowDepth[GMP_SnowDepth['Measurement Tool'] =='PR']
```

```python
# group data by the 'snow pit ID' and show {summary statistics} of 'snow depth' around each snow pit
GMP_SnowDepth.groupby(['PitID'])[['Depth (cm)']].describe().round(1)
```

I thought there are 150 snow pits. How come there are only 148 of them with observations. Some pits have records one time.

```python
#Show pits with less than 50 snow depth observations.

#pitObservationCount 
pitObservationCount = GMP_SnowDepth.groupby(['PitID'])[['Depth (cm)']].count().rename(columns = {'Depth (cm)': 'count'})
pitObservationCount[pitObservationCount['count'] <= 50].sort_values(by ='count')
```

8 snow pits have just one observation....?? What's the naming convention for the FL2A and GML

```python
# Select records associated with a certain Pit Ids

PitID_8NP = GMP_SnowDepth[GMP_SnowDepth['PitID'] == '8N9']
print('The shape of the 8NP is: ', PitID_8NP.shape)
PitID_8NP
```

Since the code convention is 8N9; it is a dense vegetation with intermediate snow pack (90-122 cm) is expected.

## Plotting

```python
# Select records associated with a Pit ruler
pit_ruler_depth = SnowDepth[SnowDepth['Measurement Tool'] == 'PR']

#plot the pit ruler depth
ax = pit_ruler_depth.plot(x='Easting', y='Northing', c='Depth (cm)', kind='scatter', alpha=0.7, colorbar=True, colormap='PuBu', legend=True)
ax.set_title('Grand Mesa Pit Ruler Depths')
ax.set_xlabel('Easting [m]')
ax.set_ylabel('Northing [m]')
plt.show()
```
